```python
filename_in = 'GIOVANNI-g4.timeAvgMap.FLDAS_NOAH01_C_GL_MA_001_Evap_tavg.20200501-20200531.180W_60S_180E_90N.csv'
year = 2022
month = 'April'
title = 'Time Averaged Map of Anomaly of Total monthly ET 0.1 deg. FLDAS_NOAH01_C_GL_MA_001'
```

```python
################################################################
# Purpose: Read CSV satellite data 
# Author: Aster Volta
# Date: September 19, 2022
################################################################

# Set up the libraries and modules
#
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import geopandas as gpd

import rasterio
import seaborn as sns
```

```python
df_et_a = pd.read_csv(filename_in, header=None)
```

```python
# HeatMap
ax = plt.subplots(figsize=(16,8))         # Sample figsize in inches
ax = sns.heatmap(df_et_a, xticklabels=False, yticklabels=False, center=0, cmap="PRGn", #vmin=-7e-6, vmax=7e-6,
                       cbar_kws={'label': '$kg\;m^{-2}\;s^{-1}$'}) 
# Add Date
ax = plt.title(title+f' for {month} {year}')
```
